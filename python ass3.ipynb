```python
import numpy as np
import pandas as pd

data = {
    'Name': ['shazia', 'noor', 'amna', 'laiba'],
    'Age': [10, 20, 30, 40],
    'Salary': [20000, 30000, 40000, 50000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

numeric_data = df.select_dtypes(include=['int64', 'float64']).to_numpy()

print("\nArray Properties:")
print("Shape:", numeric_data.shape)
print("Data type (dtype):", numeric_data.dtype)
print("Number of dimensions (ndim):", numeric_data.ndim)

print("\nVectorized operations:")
print("Mean (column-wise):", np.mean(numeric_data, axis=0))
print("Standard Deviation (column-wise):", np.std(numeric_data, axis=0))

addition_result = numeric_data + 10
multiplication_result = numeric_data * 2

print("\nElement-wise Addition (+10):")
print(addition_result)

print("\nElement-wise Multiplication (*2):")
print(multiplication_result)
```

```python

import pandas as pd
import numpy as np

data = {
    'Name': ['Shazia', 'Noor', 'Amna', 'Laiba', 'Falak', 'Bisma', np.nan],
    'Age': [22, 28, 35, np.nan, 40, 30, 25],
    'Salary': [45000, 52000, np.nan, 75000, 82000, 61000, 50000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'Finance', 'HR', 'HR']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

print("\nBasic Info:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())
print("\nDataFrame Info:")
print(df.info())

print("\nHandling Missing Values:")
df.fillna({'Age': df['Age'].mean(), 'Salary': df['Salary'].median()}, inplace=True)
df.dropna(subset=['Name'], inplace=True)
print(df)

print("\nFixing Data Types:")
df['Age'] = df['Age'].astype(int)
df['Department'] = df['Department'].astype('category')
print(df.dtypes)

print("\nRemoving Duplicates (if any):")
df = df.drop_duplicates()
print(df)

print("\nEmployees older than 30:")
print(df[df['Age'] > 30])

print("\nSelecting columns (Name and Salary) using .loc:")
print(df.loc[:, ['Name', 'Salary']])

print("\nSelecting first 3 rows using .iloc:")
print(df.iloc[0:3])

print("\nGrouping by Department:")
grouped = df.groupby('Department', observed=True).agg({
    'Salary': ['mean', 'min', 'max', 'count'],
    'Age': ['mean']
})
print(grouped)

dept_data = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance'],
    'Manager': ['Areeba', 'Sana', 'Iqra']
})

merged_df = pd.merge(df, dept_data, on='Department', how='left')
print("\nMerged DataFrame (with Manager info):")
print(merged_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [150, 200, 250, 300, 280, 350]

categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 12]

np.random.seed(0)
ages = np.random.randint(18, 60, 100)
height = np.random.randint(150, 200, 100)
weight = np.random.randint(50, 100, 100)

plt.figure(figsize=(8,5))
plt.plot(months, sales, marker='o', linestyle='-', color='blue', label='Sales')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
plt.bar(categories, values, color='orange', edgecolor='black')
plt.title('Category Value Comparison')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

plt.figure(figsize=(8,5))
plt.hist(ages, bins=10, color='green', edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(height, weight, color='red', marker='x', alpha=0.7, label='Height vs Weight')
plt.title('Height vs Weight Scatter Plot')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

colors = ['gold', 'lightblue', 'lightgreen', 'pink']
plt.figure(figsize=(6,6))
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True)
plt.title('Category Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample Data
np.random.seed(0)
data = pd.DataFrame({
    'Age': np.random.randint(18, 60, 100),
    'Height': np.random.randint(150, 200, 100),
    'Weight': np.random.randint(50, 100, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Category': np.random.choice(['A', 'B', 'C'], 100)
})

# 1. Heatmap: Correlation between numerical columns
plt.figure(figsize=(8,6))
corr = data[['Age','Height','Weight']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('1. Correlation Heatmap')
plt.show()

# 2. Pairplot: Multi-variable relationships
sns.pairplot(data[['Age','Height','Weight','Gender']], hue='Gender', palette='Set2')
plt.suptitle('2. Pairplot of Age, Height & Weight by Gender', y=1.02)
plt.show()

# 3. Violin Plot: Distribution comparison
plt.figure(figsize=(8,5))
sns.violinplot(x='Category', y='Weight', data=data, palette='pastel', inner='box')
plt.title('3. Weight Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Weight')
plt.show()
```
